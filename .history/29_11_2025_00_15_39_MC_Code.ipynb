```python
import pandas as pd
url_player = "https://drive.google.com/uc?export=download&id=1Mw9vW0hjTJwRWx0bDXiSpYsO3gKogaPz"
url_session = "https://drive.google.com/uc?export=download&id=14O91N5OlVkvdGxXNJUj5jIsV5RexhzbB"
player_data = pd.read_csv(url_player)
session_data = pd.read_csv(url_session)
player_data
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsRegressor

player_filtered = player_data.loc[:, ["played_hours", "gender", "age"]]

mc_train, mc_test = train_test_split(
    player_filtered, train_size = 0.75
)

mc_preprocessor = make_column_transformer(
    (StandardScaler(), ["age"]),
)

mc_pipeline = make_pipeline(mc_preprocessor, KNeighborsRegressor())
param_grid = {
    "kneighborsregressor__n_neighbors":range(1,201,3),
}

mc_gridsearch = GridSearchCV(
    estimator = mc_pipeline,
    param_grid = param_grid,
    cv = 5,
    scoring = "neg_root_mean_squared_error",
)
mc_gridsearch.fit(
    mc_train[["age"]],
    mc_train["played_hours"]
)
mc_results = pd.DataFrame(mc_gridsearch.cv_results_)
mc_results
```
